```python
import os
import pandas as pd
from PIL import Image
import pytesseract

# 1. Tesseract path eka set karanna
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# 2. Folder paths
input_hateful = "raw_hateful"
input_non_hateful = "raw_non_hateful"
output_folder = "processed_dataset"

# Output folder eka nattam hadanna
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

data_list = []

def process_memes(folder, label, prefix):
    if not os.path.exists(folder):
        print(f"Error: '{folder}' folder eka hoyaganna na!")
        return

    # Image files tika list karaganna
    files = [f for f in os.listdir(folder) if f.lower().endswith(('.jpg', '.png', '.jpeg', '.webp'))]
    print(f"Processing {len(files)} {label} images...")
    
    for i, filename in enumerate(files, 1):
        try:
            # Photo eka open karala standardized size ekata ($512\times512$) resize karanna
            img_path = os.path.join(folder, filename)
            img = Image.open(img_path).convert('RGB')
            img = img.resize((512, 512)) 
            
            # Aluth nama dila save karanna (hate_1.jpg, non_hate_1.jpg wage)
            new_name = f"{prefix}_{i}.jpg"
            save_path = os.path.join(output_folder, new_name)
            img.save(save_path)
            
            # Tesseract OCR use karala Sinhala text extract karanna
            text = pytesseract.image_to_string(img, lang='sin').replace('\n', ' ').strip()
            
            data_list.append({
                'Image_Name': new_name, 
                'Label': label, 
                'Extracted_Text': text
            })
            
            if i % 10 == 0: print(f"Progress: {i}/{len(files)}")
            
        except Exception as e:
            print(f"Error processing {filename}: {e}")

# --- START THE PRE-PROCESSING ---
process_memes(input_hateful, "Hateful", "hate")
process_memes(input_non_hateful, "Non-Hateful", "non_hate")

# Excel file ekakata data save karamu
if data_list:
    df = pd.DataFrame(data_list)
    # Shuffle the data
    df = df.sample(frac=1).reset_index(drop=True)
    df.to_excel("meme_dataset_final.xlsx", index=False)
    print("\n--- JAYAWEWA! ---")
    print(f"1. Photos okkoma '{output_folder}' folder ekata save una.")
    print(f"2. Data okkoma 'meme_dataset_final.xlsx' file ekata save una.")
else:
    print("Folders wala photos thiyenawada kiyala ayeth check karanna.")
```

```python
import os
import pytesseract
import pandas as pd
from PIL import Image
from tqdm import tqdm

# 1. Configure Tesseract OCR path (verify this path on your system)
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

# 2. Define input and output paths
input_folder = "processed_dataset"
output_file = "meme_dataset_tesseract.xlsx"

data_list = []

# 3. Get a list of all supported image files in the input folder
files = [f for f in os.listdir(input_folder) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

print(f"Extracting text from {len(files)} images using Tesseract OCR...")

# 4. Iterate through each image and extract text
for filename in tqdm(files):
    try:
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)
        
        # Extract text using both Sinhala (sin) and English (eng) language models
        text = pytesseract.image_to_string(img, lang='sin+eng')
        
        # Infer label from filename (e.g., contains 'hate' but not 'non_hate')
        label = "Hateful" if "hate" in filename and "non_hate" not in filename else "Non-Hateful"
        
        data_list.append({
            'Image_Name': filename,
            'Label': label,
            'Extracted_Text': text.strip()
        })
    except Exception as e:
        print(f"Error processing {filename}: {e}")

# 5. Convert extracted data to a pandas DataFrame
df = pd.DataFrame(data_list)

# --- Text Preprocessing (Clean the extracted raw text) ---
def simple_clean(text):
    import re
    if not text: 
        return ""
    
    # Remove numbers
    text = re.sub(r'[0-9]+', '', text) 
    # Remove special characters and symbols
    text = re.sub(r'[|\[\]\{\}_*+=/\\<>@#$]', '', text) 
    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text).strip() 
    
    return text

# Apply the cleaning function to the extracted text column
df['Extracted_Text'] = df['Extracted_Text'].apply(simple_clean)

# Save the final structured data to an Excel file
df.to_excel(output_file, index=False)

print(f"\nExtraction complete! Data saved successfully to '{output_file}'.")
print("You can now open the file to manually review and refine the text.")
```

```python
import pandas as pd
import re

# 1. File eka load karamu
input_file = "meme_dataset_final.xlsx"
df = pd.read_excel(input_file)

print(f"Total Rows: {len(df)}")

# --- Cleaning Function ---
def clean_text(text):
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # 1. Numbers ain karanna
    text = re.sub(r'[0-9]+', '', text)
    
    # 2. Special Symbols ain karanna
    text = re.sub(r'[|\[\]\{\}_*+=/\\<>@#$]', '', text)
    
    # 3. Multiple spaces ain karala thani space ekak karanna
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Cleaning function eka apply karamu
# (Original 'Extracted_Text' column ekama update karanawa)
df['Extracted_Text'] = df['Extracted_Text'].apply(clean_text)

# --- NOTE: Api dan Empty Rows ain karanne NA ---
# User manual fill karana nisa, okkoma rows thiya gannawa.

# Save New File
output_file = "meme_dataset_cleaned_manual.xlsx"
df.to_excel(output_file, index=False)

print(f"\n✅ Wade Goda! Clean karapu file eka '{output_file}' namal save una.")
print(f"Total Rows: {len(df)} (Kisima row ekak makune na).")
print("Dan eka open karala, his (empty) than walata galapena text type karanna.")
```
